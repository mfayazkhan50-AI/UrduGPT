```python
from datasets import load_dataset
```

```python
dataset = load_dataset("abdullah693/adaption-urdu-edu-cultural-reasoning")
print(dataset['train'][0])
```

```python
urdu_data = dataset['train'].filter(lambda x: x['lang'] == 'ur')
print(f"Urdu rows: {len(urdu_data)}")
print(urdu_data[0])
```

```python
def format_data(row):
  return {
      "text": f"### Instruction:\n{row['prompt']}\n\n### Response:\n{row['enhanced_completion']}"
  }

formatted_data = urdu_data.map(format_data)
print(formatted_data[0]['text'])
```

```python
import torch
print(torch.cuda.is_available())
print(torch.cuda.get_device_name(0))
```

```python
from unsloth import FastLanguageModel

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = 'unsloth/llama-3.2-3b-instruct',
    max_seq_length = 2040,
    load_in_4bit = True
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=[
        'q_proj', 'k_proj', 'v_proj', 'o_proj'
    ],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = 'none'
)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments
```

```python
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=formatted_data,
    dataset_text_field="text",
    max_seq_length = 2024,
    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=4,
        warmup_steps=10,
        max_steps=100,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=10,
        output_dir='outputs',
        save_strategy='no'
    ),

)
```

```python
trainer.train()
```

```python
model.save_pretrained('urdu_model')
tokenizer.save_pretrained('urdu_model')
print("Model Saved")
```
